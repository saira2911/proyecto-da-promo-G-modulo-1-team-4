```python
import random

class Juegos:
    def __init__(self):
        pass

    def piedra_papael_tijera(self):

        #ponia que jugara un jugador contra otro pero, de la forma en que tenemos que a√±adir los inputs ahora mismo ambos jugadores verian sus respuestas
        #de todas formas si el cliente lo quiere para 2 jugadores se puede modificar de forma sencilla usando un input en lugar de random en la fila 23
        #palabra_secreta = input("Jugador 1, introduce piedra, papael o tijera")
        lista = ["piedra","papel","tijera"]

        rondas = 3
        #aqui se almacenan M (maquina) o U (usuaria) segun quien gane cada ronda para luego hacer recuento
        resultado = []

        print("Vamos a jugar a piedra, papel, tijera, a ver si puedes conmigo! üòè")

        while rondas > 0 and resultado.count("M") < 2 and resultado.count("U") < 2: ## he modificado para que cuando alguno lleve 2 rondas ganadas se salga del bucle y no espere a resolver la tercera ronda
            
            usuario=input("Elige piedra, papel, tijera").lower()
            maquina = random.choice(lista)
            
            if usuario == "piedra":
                if maquina == "piedra":
                    print(f"Vaya! quedamos empate üòÇ, yo tambien eleg√≠ {maquina}  ") ## johanna a√±ade piedra
                elif maquina == "papel":
                    print(f"ü§ñ Yo gano üòé he elegido {maquina} üßæ")
                    rondas -= 1
                    resultado.append("M")
                else: #tijera
                    print(f"Bien hecho!! ü•≥ has ganado! üèÜ yo eleg√≠ {maquina} ‚úÇ ")
                    rondas -= 1  
                    resultado.append("U")

            elif usuario == "papel":
                if maquina == "piedra":
                    print(f"Ah! üò≤ Ganaste esta ronda! üòÅ yo eleg√≠ {maquina} ") ## johanna a√±ade piedra
                    rondas -= 1
                    resultado.append("U")
                elif maquina == "papel":
                    print(f"Empate! üôÑ yo tambien eleg√≠ {maquina} üßæ ")
                else: #tijera
                    print(f"ü§ñ Ganeeee! ü§© He elegido {maquina} ‚úÇ ")
                    rondas -= 1
                    resultado.append("M")
            
            elif usuario == "tijera":
                if maquina == "piedra":
                    print(f"ü§ñ He ganado! ü§≠ mi eleccion ha sido {maquina} ") ## johanna a√±ade piedra
                    resultado.append("M")
                    rondas -= 1  
                elif maquina == "papel":
                    print(f"Bien jugado! Ganaste! üéâ Yo eleg√≠ {maquina} üßæ ")
                    rondas -= 1
                    resultado.append("U")
                else: #tijera
                    print(f"Quedamos empatadas! ü§î Yo tambien eleg√≠ {maquina} ‚úÇ ")
                    
            else:
                print("Uy! Te has equivocado! ‚ùå Por favor elige: piedra, papel o tijera")
                continue

        if resultado.count("U") >= 2:
            print("Me has ganado la partida!! ü•á Quiero la revancha! üí™")

        elif resultado.count("M") >= 2:
            print("ü§ñ He ganado esta partida! Juguemos otra vez! üòã")

    def ahorcado (self):

        #ponia que jugara un jugador contra otro pero, de la forma en que tenemos que a√±adir los inputs ahora mismo ambos jugadores verian sus respuestas
        #de todas formas si el cliente lo quiere para 2 jugadores se puede modificar de forma sencilla usando un input en lugar de random en la fila 79
        #palabra_secreta = input("Jugador 1, introduce la palabra secreta")

        palabras = ["caba√±a", "gato", "analista", "sandia", "monitor"]

        palabra_secreta = random.choice(palabras)

        #transformamos la palabra secreta en oculta con *
        palabra_oculta = "*" * len(palabra_secreta)

        #a√±adimos las vidas
        vidas = 6
        #mientras que los strings no coincidan y no se hayan agotado las vidas
        while palabra_oculta != palabra_secreta and vidas > 0:
            letra = input("Introduzca otra letra").lower()

            print(f"La letra que has introducido es {letra}")

            #si la letra que da la usuaria esta en la palabra secreta
            if letra in palabra_secreta:
                #para cada posicion (indice) de la palabra secreta
                for i in range(len(palabra_secreta)):
                    #si alguna letra de palabra secreta coincide con la letra de la usuaria
                    if palabra_secreta[i]==letra:
                        #reemplazamos en la palabra oculta el * correspondiente por la letra de esta forma ya que con replace encontraba el primer * y lo cambiaba sin importar la posicion
                        palabra_oculta = palabra_oculta[:i]+letra+palabra_oculta[i+1:]
                    else:
                        pass
                if palabra_oculta == palabra_secreta: 
                    print(f"Enhorabuena! ü•≥ Has acertado la palabra {palabra_secreta}") 
                    break 
                else: 
                    print(f"Has acertado una letra üéâ, la palabra esta asi: {palabra_oculta}") 
                posible_resolucion = input("Sabes cual es la palabra?responde si/no").lower()
                if posible_resolucion == "si":
                    intento_respuesta = input("dime la palabra").lower()
                    if intento_respuesta == palabra_secreta:
                        print(f"Enhorabuena! ü•≥ has acertado la palabra {palabra_secreta}")
                        break
                    else:
                        vidas -= 1
                        if vidas > 0: 
                            print(f"La palabra que has introducido es {intento_respuesta} te has equivocado üí•, sigue jugando")
                            ############aqui podemos a√±adir otra vez todo el mu√±eco para que si fallas la palabra completa ademas de decirte las vidas que te quedan te muestre el mu√±eco
                            print(f"te quedan {vidas} vidas")
                        else: 
                            print(f"La palabra {intento_respuesta} no es correcta ‚ùå y te has quedado sin vidas") 
                            print("Has perdido el juego üò≠") 
                            break 

                elif posible_resolucion == "no":
                        print(f"Vale! No pasa nada! üëç sigue jugando")
                        
            #si la letra que da el ususario no coincide con ninguna de la palabra secreta
            else:
                vidas -= 1
                #mostramos el resultado al usuario
                #para el mu√±eco: si la letra no esta en la palabra secreta tiene que imprimir cada parte del mu√±eco
                print(f"Esa letra no esta en la palabra üò£, la palabra esta asi: {palabra_oculta}")
                if vidas == 5: 
                    print("O")
                    print(f"te quedan {vidas} vidas")

                elif vidas == 4:
                    print(" O")
                    print("/")
                    print(f"te quedan {vidas} vidas")

                elif vidas == 3:
                    print(" O")
                    print("/|")
                    print(f"te quedan {vidas} vidas")   

                elif vidas == 2:
                    print(" O")
                    print("/|\\")
                    print(f"te quedan {vidas} vidas")

                elif vidas == 1:
                    print(" O")
                    print("/|\\")
                    print("/")
                    print(f"te quedan {vidas} vidas")
                
                elif vidas == 0:
                    print(" O")
                    print("/|\\")
                    print("/ \\")
                    print(f"te quedan {vidas} vidas")

                    print("Has perdido el juego")
                    break

    def adalabers_viajeras (self):

        lista_paises = ["Francia", "Italia", "Brasil"]

        diccionario_monumentos = {"Francia" : "Torre Eiffel", "Italia" : "Coliseo", "Brasil" : "Cristo Redentor"}
        diccionario_gentilicios = {"Francia" : "Francesa", "Italia" : "Italiana", "Brasil" : "Brasile√±a"}
        diccionario_colores_banderas = {"Francia" : ["Azul", "Blanco", "Rojo"], "Italia" : ["Verde", "Blanco", "Rojo"], "Brasil": ["Verde", "Azul", "Amarillo"]}
        diccionario_capitales = {"Francia" : "Paris", "Italia" : "Roma", "Brasil" : "Brasilia"}
        diccionario_olimpiadas_verano = {"Francia" : "2024", "Italia" : "1960", "Brasil" : "2016"}
        diccionario_mares = {"Francia" : ["Mar Mediterraneo", "Oceano Atlantico"], "Italia" : ["Mar Jonico", "Mar De Liguria", "Mar Tierno", "Mar Adriatico"], "Brasil" : ["Oceano Atlantico"]}
        
        vidas = 4

        print(f"Elige uno de estos paises y veremos cuanto sabes de el üòâ: {lista_paises}")

        while vidas > 0:
            pais = input(f"Elige el pais que quieres entre {lista_paises}").title()

            if pais in lista_paises:

                print("Vamos alla, ¬°mucha suerte! üí™")
                #PREGUNTA 1: la usuaria tiene que acertar el gentilicio
                print(f"Primera pregunta: ¬øCual es el gentilicio femenino de {pais}? üëÄ")
                respuesta_gentilicio = input(f"Escribe el gentilicio femenino de {pais}").title()

                if diccionario_gentilicios[pais] == respuesta_gentilicio:
                    print(f"Genial! Has acertado! El gentilicio de {pais} es {respuesta_gentilicio} ü§©")
                else:
                    vidas -= 1
                    print(f"Vaya! üò• Has dicho {respuesta_gentilicio} y la respuesta correcta es {diccionario_gentilicios[pais]}") 
                    print(f"Te quedan {vidas} vidas")
                    print("No te preocupes! Aun quedan muchas preguntas üß†")
                #PREGUNTA 2: la usuaria tiene que acertar el monumento
                print(f"Vamos a por la segunda pregunta: ¬øCual es el monumento mas emblematico de {pais}? üóø")
                respuesta_monumento = input(f"Escribe el nombre del monumento mas importante de {pais}").title()
                
                if diccionario_monumentos[pais] == respuesta_monumento:
                    print(f"Yuhu! Has acertado! üéâ El monumento mas emblematico de {pais} es {respuesta_monumento}")
                else:
                    vidas -= 1
                    print(f"Ups! üò£ Has dicho {respuesta_monumento} y la respuesta correcta es {diccionario_monumentos[pais]}")
                    print(f"Te quedan {vidas} vidas")

                    ####podemos a√±adir si quiere cambiar de pais a mitad de juego por si sabe mas sobre otro pais??
                #PREGUNTA 3: la ususaria tiene que acertar la capital
                print(f"Tercera pregunta üòè: ¬øCual es la capital de {pais}?")
                respuesta_capital = input(f"Escribe la capital de {pais}").title()
                
                if diccionario_capitales[pais] == respuesta_capital:
                    print(f"Yas Queen! üëë Has acertado! La respuesta es {respuesta_capital}")
                    print(f"Se nota que sabes mucho de {pais} üòé")
                else:
                    vidas -= 1
                    print(f"Ohhh! üòî Has dicho {respuesta_capital} y la respuesta correcta es {diccionario_capitales[pais]}")
                    print(f"Te quedan {vidas} vidas")

                print("Vamos!! Ya casi lo tienes üî•")
                #PREGUNTA 4: la usuaria tiene que acertar todos los colores de la bandera
                print(f"Cuarta pregunta: ¬øCuales son los colores de la bandera de {pais}?")
                respuesta_bandera = input(f"Escribe los colores de la bandera de {pais} separados por coma").title()

                todos_colores = True
                for color in respuesta_bandera.split(","):
                    if color not in diccionario_colores_banderas[pais]:
                        todos_colores = False
                        print(color)

                if todos_colores == True:
                    print(f"WOW! ü•≥ Has acertado!!! Los colores de la bandera de {pais} son {', '.join(respuesta_bandera.split(','))}")
                    print(f"Se ve que hay mucho color en tu vida! üåà")
                else:
                    vidas -= 1
                    if vidas > 0:
                        print(f"Ups! üò£ Has dicho {', '.join(respuesta_bandera.split(','))} y la respuesta correcta es {', '.join(diccionario_colores_banderas[pais])}")
                        print(f"Te quedan {vidas} vidas")
                    else:
                        print(f"Vaya! Has dicho {', '.join(respuesta_bandera.split(','))} y la respuesta correcta es {', '.join(diccionario_colores_banderas[pais])}")
                        print(f"GAME OVER!‚ùå Te has quedado sin vidas pero puedes volver a intentarlo cuando quieras üòã")
                        break
                #PREGUNTA 5: la usuaria tiene que acertar el a√±o
                print(f"Vamos a por la quinta pregunta: ¬øEn que a√±o se celebraron las olimpiadas de verano en {pais}?")
                respuesta_olimpiadas = input(f"Escribe el a√±o en que se celebraron las olimpiadas de verano en {pais}").title()
                
                if diccionario_olimpiadas_verano[pais] == respuesta_olimpiadas:
                    print(f"Siii! Has acertado! üéâ El a√±o de las olimpiadas en {pais} es {respuesta_olimpiadas} ü•á")
                else:
                    vidas -= 1
                    if vidas > 0:
                        print(f"Uff casi aciertas! üò£ Has dicho {respuesta_olimpiadas} y la respuesta correcta es {diccionario_olimpiadas_verano[pais]}")
                        print(f"Te quedan {vidas} vidas")
                    else:
                        print(f"Vaya! Has dicho {respuesta_olimpiadas} y la respuesta correcta es {diccionario_olimpiadas_verano[pais]}")
                        print(f"GAME OVER!‚ùå Te has quedado sin vidas pero puedes volver a intentarlo cuando quieras üòã")
                        break
                #PREGUNTA 6: tenemos una lista de mares y oceanos y la ususaria solo tiene que poner uno
                print(f"Ultima pregunta: ¬øPuedes decirme algun mar u oceano que ba√±e las costas de {pais}?")
                respuesta_aguas = input(f"Escribe algun mar u oceano de {pais}").title()

                if respuesta_aguas in diccionario_mares[pais]:
                    print(f"Lo has hecho la mar de bien!! üåä El {respuesta_aguas} es uno de los que ba√±a {pais}") 
                    print(f"Se ve que eres una Adalaber Viajera!! üõ´ Te mereces un viaje gratis a {pais} üõ¨")
                else:
                    vidas -= 1
                    if vidas > 0:
                        print(f"Ups! üò£ Has dicho {respuesta_aguas} y la respuesta correcta es {', '.join(diccionario_mares[pais])}")
                        print(f"Te quedan {vidas} vidas")
                    else:
                        print(f"Vaya! Has dicho {respuesta_aguas} y la respuesta correcta es {', '.join(diccionario_mares[pais])}")
                        print(f"GAME OVER!‚ùå Te has quedado sin vidas pero puedes volver a intentarlo cuando quieras üòã")
                        break
                    
                if vidas == 4:
                    print("Enhorabuena!! üéä Has ganado el juego sin perder ninguna vida! ü§Ø Se nota que sabes much√≠simo de geografia ü•≥")
                    break
                elif vidas < 4 and vidas >0:
                    print("Enhorabuena!! üéä Has ganado el juego! üòé Se nota que sabes mucho de geografia ü•≥")
                    break

            else:
                print(f"Por favor elige un pais de la lista {lista_paises}")        

    
        
    
```
